## Import packages

```python
import pandas as pd
import matplotlib .pyplot as plt
import seaborn as sns
```

## Import DataFrame

```python
df=pd.read_csv("carprices.csv")
```

```python
df
```

```python
df.info()
```

```python
# Create a figure with two subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 7, figsize=(12, 6))

# Plot 'fueltype' value counts on the first subplot
df["fueltype"].value_counts().plot(kind="bar", ax=axes[0], title='Fuel Type')
df["aspiration"].value_counts().plot(kind="bar", ax=axes[1], title='aspiration')
df["doornumber"].value_counts().plot(kind="bar", ax=axes[2], title='doornumber')
df["carbody"].value_counts().plot(kind="bar", ax=axes[3], title='carbody')
df["drivewheel"].value_counts().plot(kind="bar", ax=axes[4], title='drivewheel')
df["enginetype"].value_counts().plot(kind="bar", ax=axes[5], title='enginetype')
df["cylindernumber"].value_counts().plot(kind="bar", ax=axes[6], title='cylindernumber')
df["fuelsystem"].value_counts().plot(kind="bar", ax=axes[6], title='fuelsystem')

# Display the plots
plt.tight_layout()
plt.show()
```

## Removing unnecessary columns

```python
predictors=df.drop(columns=["symboling","car_ID","CarName","aspiration","carheight","peakrpm","compressionratio","carbody","enginetype","fuelsystem"],axis=1)
```

```python
predictors
```

## Check Value for each nonInt Rows

```python
predictors.info()
```

```python
# List of columns to plot
columns_to_plot = ["fueltype", "doornumber","drivewheel","enginelocation","cylindernumber"]

# Create a figure with subplots
fig, axes = plt.subplots(1, len(columns_to_plot), figsize=(12, 6))

# Loop through each column and plot the value counts
for i, col in enumerate(columns_to_plot):
    predictors[col].value_counts().plot(kind="bar", ax=axes[i], title=col.capitalize())

# Display the plots
plt.tight_layout()
plt.show()
```

## Represent Rows values as Integers

```python
# Dictionary to map door number words to corresponding integers
column_mappings = {
    'fueltype':{"gas":1, "diesel":2},
    'doornumber':{"four":4,"two":2},
    'drivewheel' : {"fwd" :1,"rwd": 2,"4wd":3},
     'enginelocation': {"front": 1, "rear": 2},
    'cylindernumber': {"four": 4, "six": 6, "five": 5, "eight": 8, "two": 2, "three": 3, "twelve": 12}
}

predictors['fueltype'] = predictors['fueltype'].map(column_mappings['fueltype']).astype('category')
predictors['doornumber'] = predictors['doornumber'].map(column_mappings['doornumber']).astype('category')
predictors['drivewheel'] = predictors['drivewheel'].map(column_mappings['drivewheel']).astype('category')
predictors['enginelocation'] = predictors['enginelocation'].map(column_mappings['enginelocation']).astype('category')
predictors['cylindernumber'] = predictors['cylindernumber'].map(column_mappings['cylindernumber']).astype('category')
```

```python
predictors.info()
```

## Split the Data

```python
from sklearn.model_selection import train_test_split
```

```python
predictors
```

```python
# X is for Predictors, calling the whole table and dropping price
X=predictors.drop(columns=["price"],axis=1)
Y=predictors['price']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=.2, random_state=1)
```

## Train the model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
predictions=lm.predict(X_test)
train_predict=lm.predict(X_train)
```

```python
results_df = pd.DataFrame({
    'Actual': Y_test,
    'Predicted': predictions
})

sns.lmplot(x='Actual', y='Predicted', data=results_df, height=6, aspect=1.2)
plt.title('Linear Regression Plot: Actual vs Predicted')
plt.show()
```

```python
from sklearn import metrics
error_score_linear = metrics.r2_score(Y_test,predictions)
print("R Squared error:", error_score_linear)
```

## TRAINING USING LASSO REGRESSION

```python
from sklearn.linear_model import Lasso
```

```python
lrm=Lasso()
```

```python
lrm.fit(X_train,Y_train)
```

```python
lrmpredictions=lrm.predict(X_test)
```

```python
plt.scatter(x=lrmpredictions,y=Y_test)
plt.xlabel('Predictions')
plt.ylabel('Actual Test')
plt.title('Lasso Regression Plot: Actual vs Predicted')
plt.show()
```

```python
from sklearn import metrics
error_score_lasso = metrics.r2_score(Y_test,lrmpredictions)
print("LASSO R Squared error:", error_score_lasso)
```

```python
# X is for Predictors, calling the whole table and dropping price
X=predictors[["horsepower","enginesize","curbweight"]]
Y=predictors['price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=.2, random_state=2021)
```

```python
Y_train
```

```python
from xgboost import XGBRegressor
xgb_model=XGBRegressor()
```

```python
search_space ={
    "n_estimators":[100,200,500],
    "max_depth":[3,6,9],
    "gamma":[0.01,0.1],
    "learning_rate":[0.001,0.01,0.1,1]
}
```

```python
from sklearn.model_selection import GridSearchCV
GS=GridSearchCV(estimator= xgb_model, 
                param_grid= search_space,
                scoring= ["r2","neg_root_mean_squared_error"],
                refit= "r2",
                cv = 5,
                verbose = 4)
```

```python
GS.fit(X_train,Y_train)
```

```python
print(GS.best_estimator_)
```

```python
print(GS.best_params_)
```

```python
gspredict=GS.predict(X_test)
```

```python
plt.scatter(x=gspredict,y=Y_test)
plt.xlabel('Predictions')
plt.ylabel('Actual Test')
plt.title('XGBOOST Plot: Actual vs Predicted')
plt.show()
```

```python
from sklearn import metrics
error_score_xgb = metrics.r2_score(Y_test,gspredict)
print("XGB R Squared error:", error_score_xgb)
```

```python


print("Linear R Squared error:", error_score_linear)
print("LASSO R Squared error:", error_score_lasso)
print("XGB Squared error:", error_score_xgb)
```
